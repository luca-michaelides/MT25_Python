```python
# Problem 20

def factorial(n: int):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

def sum_of_digits(n: int):
    k = len(str(n))
    sum = 0
    for i in range(k):
        sum += int(str(n)[i])
    return sum

sum_of_digits(factorial(100))
```

```python
# Problem 21

def proper_divisors(n: int):
    if n == 1:
        return []
    elif (n == 2) or (n == 3):
        return [1]
    else:
        divisors = [1]
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            divisors.append(n//i)
    if n**0.5 == int(n**0.5):
        del divisors[-1]
    return divisors

def amicable_numbers_sum(n: int):
    result = 0
    idx = list(range(1,n))
    amicable = []
    while idx:
        i = idx[0]
        d_n = sum(proper_divisors(i))
        if (i == sum(proper_divisors(d_n))) & (i != d_n):
            result += (i+d_n)
            idx.remove(d_n)
            amicable.append(i)
            amicable.append(d_n)
        del idx[0]  
    return result

amicable_numbers_sum(10_000)
```
