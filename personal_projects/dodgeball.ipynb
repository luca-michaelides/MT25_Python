#### Dodgeball - Wincent Assessment ####

```python
import json
import numpy as np

with open('test.in.json', 'r') as file:
    submission_data = json.load(file)

def dodgeball(n_p, start_d, start_p, p_coords):
    directions = {
        'N':    (0,1), 
        'NE':   (1,1), 
        'E':    (1,0), 
        'SE':  (1,-1), 
        'S':   (0,-1), 
        'SW': (-1,-1), 
        'W':   (-1,0), 
        'NW':  (-1,1)
    }
    directions_order = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
    throw_count = 0
    current_player = start_p
    current_direction_idx = directions_order.index(start_d)
    remaining_players = list(range(1,n_p + 1))
    remaining_players.remove(current_player)
    while remaining_players:
        rotation_count = 0
        found_receiver_in_this_turn = False
        while rotation_count < len(directions_order):
            current_direction_idx = (current_direction_idx + 1) % len(directions_order)
            current_direction_coords = directions[directions_order[current_direction_idx]]
            rotation_count += 1
            min_scale = float('inf')
            receiver = None
            for player in remaining_players:
                dx = p_coords[player - 1][0] - p_coords[current_player - 1][0]
                dy = p_coords[player - 1][1] - p_coords[current_player - 1][1]
                scale = float('inf')
                if current_direction_coords[0] == 0:
                    if dx == 0:
                        scale = dy / current_direction_coords[1]
                elif current_direction_coords[1] == 0:
                    if dy == 0:
                        scale = dx / current_direction_coords[0]
                else:
                    x_scale = dx / current_direction_coords[0]
                    y_scale = dy / current_direction_coords[1]
                    if abs(x_scale - y_scale) < 1e-9:
                        scale = x_scale
                if scale > 0 and scale < min_scale:
                    min_scale = scale
                    receiver = player
            if receiver is not None:
                throw_count += 1
                current_player = receiver
                remaining_players.remove(current_player)
                current_direction_idx = (current_direction_idx + 4) % len(directions_order)
                found_receiver_in_this_turn = True
                break 
        if not found_receiver_in_this_turn:
            return throw_count, current_player            
    return throw_count, current_player

def test(data):
    result = []
    for i in range(len(data)):
        n_players = len(data[i]["players"])
        start_dir = str(data[i]["startingDirection"])
        start_player = int(data[i]["startingPlayer"])
        player_coords = np.zeros((n_players, 2), dtype=int)
        for j in range(n_players):
            player_coords[j][0] = int(data[i]["players"][j][0])
            player_coords[j][1] = int(data[i]["players"][j][1])
        value = dodgeball(n_players, start_dir, start_player, player_coords)
        result.append(f"{value[0]} {value[1]}")
    return "\n".join(result)

output = test(submission_data) + '\n'

with open('output_file.out', 'w', encoding='utf-8') as file:
    file.write(output)
```
